# 一、提出问题

在泰坦尼克号事件中什么样的人生存的几率更高

# 二、理解数据

# 2.1 导入数据

```python
# 从csv文件中导入数据
import numpy as np
import pandas as pd
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
print(train.shape)
print(test.shape)
```

```python
# 查看训练数据集
train.head()
```

```python
# 查看测试数据集
test.head()
```

```python
# 合并两个数据集
df = pd.concat([train, test], ignore_index=True)
df.head()
```

```python
# 查看df数据集每一列的信息
df.info()
```

```python
# 查看df数据集描述性信息
df.describe()
```

# 三、数据预处理

# 3.1 缺失值处理

```python
# 年龄Age缺失值用平均值填充
df['Age'].fillna(df['Age'].mean(), inplace=True)
# 船舱缺失值较多，用‘U’来填充，表示未知
df['Cabin'].fillna('U', inplace=True)
# 船票价格用平均值填充
df['Fare'].fillna(df['Fare'].mean(), inplace=True)
```

```python
# 查看登船港口Embarked频数最大的是哪一个
df['Embarked'].value_counts()
```

```python
# 登船港口Embarked仅缺失2条，用频率最高的填充
df['Embarked'].fillna('S', inplace=True)
```

```python
# 查看新的数据集
df.head()
```

```python
# 查看df每一列是否正常
df.info()
```

# 3.2 特征提取

## 3.2.1年龄Age

```python
'''3岁以下为baby；3-18岁为child；18-30岁为young；30-50岁为middle；大于50岁为old
'''
# 存放新的年龄特征
df['Age_B'] = df['Age'].map(lambda x : 1 if x<3 else 0)
df['Age_C'] = df['Age'].map(lambda x : 1 if 3<=x<18 else 0)
df['Age_Y'] = df['Age'].map(lambda x : 1 if 18<=x<30 else 0)
df['Age_M'] = df['Age'].map(lambda x : 1 if 30<=x<50 else 0)
df['Age_O'] = df['Age'].map(lambda x : 1 if x>=50 else 0)
df.head()
```

### 3.2.2船舱Cabin

```python
# 将船舱数据转化为One-hot编码
df['Cabin'] = df['Cabin'].map(lambda x : x[0])
Cabindf = pd.get_dummies(df['Cabin'], prefix='Cabin')
Cabindf.head()
```

### 3.2.3 登船港口Embarked

```python
# 将登船港口转化为One-hot编码
Embarkeddf = pd.get_dummies(df['Embarked'], prefix='Embarked')
Embarkeddf.head()
```

### 3.2.4 名称Name

```python
# 查看Name信息
df['Name'][:5]
```

```python
'''
Name中包含了一些用以区分身份的信息，定义以下几种头衔类别：
Officer政府官员
Royalty王室（皇室）
Mr已婚男士
Mrs已婚妇女
Miss年轻未婚女子
Master有技能的人/教师
'''
# 对Name进行分类
title_mapDict = {
                    "Capt":       "Officer",
                    "Col":        "Officer",
                    "Major":      "Officer",
                    "Jonkheer":   "Royalty",
                    "Don":        "Royalty",
                    "Sir" :       "Royalty",
                    "Dr":         "Officer",
                    "Rev":        "Officer",
                    "the Countess":"Royalty",
                    "Dona":       "Royalty",
                    "Mme":        "Mrs",
                    "Mlle":       "Miss",
                    "Ms":         "Mrs",
                    "Mr" :        "Mr",
                    "Mrs" :       "Mrs",
                    "Miss" :      "Miss",
                    "Master" :    "Master",
                    "Lady" :      "Royalty"
                    }
```

```python
# 从Name中获取称谓title
titledf = pd.DataFrame()
def getTitle(name):
    str1 = name.split(',')[1]
    str2 = str1.split('.')[0]
    str3 = str2.strip()
    return str3
titledf['Title'] = df['Name'].map(getTitle)
titledf[:5]
```

```python
# 将title转化为One-hot编码
titledf['Title'] = titledf['Title'].map(title_mapDict)
titledf = pd.get_dummies(titledf['Title'])
titledf.head()
```

### 3.2.5同代直系亲属数（Parch）与不同代直系亲属数（SibSp）

```python
'''
这里新定义一个FamilySize
FamilySize = Parch + SibSp + 1(自己)
'''
# 存放家庭信息
familydf = pd.DataFrame()
familydf['FamilySize'] = df['Parch'] + df['SibSp'] + 1
familydf.head()
```

```python
'''
家庭类别：
小家庭Family_Single：家庭人数=1
中等家庭Family_Small: 2<=家庭人数<=4
大家庭Family_Large: 家庭人数>=5
'''
# 对家庭大小进行分组
familydf[ 'Family_Single' ] = familydf[ 'FamilySize' ].map( lambda s : 1 if s == 1 else 0 )
familydf[ 'Family_Small' ]  = familydf[ 'FamilySize' ].map( lambda s : 1 if 2 <= s <= 4 else 0 )
familydf[ 'Family_Large' ]  = familydf[ 'FamilySize' ].map( lambda s : 1 if 5 <= s else 0 )
familydf.head()
```

### 3.2.6 客舱等级Pclass

```python
# 将客舱等级Pclass转化为One-hot编码
Pclassdf = pd.DataFrame()
Pclassdf = pd.get_dummies(df['Pclass'], prefix='Pclass')
Pclassdf.head()
```

### 3.2.7 性别Sex

```python
# 将性别转化为One-hot编码
Sexdf = pd.DataFrame()
Sexdf = pd.get_dummies(df['Sex'])
Sexdf.head()
```

### 3.2.7 合并新的数据集

```python
# 合并新的数据集
new_df = pd.concat([df, Cabindf, Embarkeddf, titledf, familydf, Pclassdf, Sexdf], axis=1)
new_df.info()
```

```python
# 去掉多余的列
new_df.drop(['Cabin', 'Embarked', 'Name', 'Sex'], axis=1,inplace=True)
new_df.info()
```

```python
# 查看数据集信息
new_df.head()
```

# 四、构建模型

## 4.1 特征选择

```python
# 查看与Survied有相关性的因素
new_df.corr()['Survived'].sort_values
```

```python
# 删除与Survived相关性较小的列
new_df.drop(['PassengerId', 'Ticket'], axis=1,inplace=True)
# 特征选择
new_df_X = new_df.drop(['Survived'], axis=1)
new_df_X.head()
```

## 4.2 创建训练数据集和测试数据集

```python
# 查看原训练数据集形状
train.shape
```

```python
sourcerow = 891
# 提出训练集特征
source_X = new_df_X.loc[:sourcerow-1, :]
# 提取训练集标签
source_y = new_df.loc[:sourcerow-1, 'Survived']
# 创建测试集特征
pred_X = new_df_X.loc[sourcerow:, :]
```

```python
# 对原始数据集进行训练
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(source_X, source_y, train_size=0.8)
```

## 4.3 机器学习算法

```python
#采用逻辑回归算法
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
#训练模型
model.fit(X_train, y_train)
```

```python
# 模型评估
model.score(X_test, y_test)
```
